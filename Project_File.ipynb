## Business Problem Statement

- What are the top 10 most common genres represented on Amazon Prime?

## Data Cleaning

```python
# Check for missing values
df.isnull().sum()
```

```python
# For the business query, dropping redunant data/Columns
df1 = df.drop(['director','cast','duration','description','date_added','title'],axis = 1)
df1
```

```python
# Fill missing values for 'rating' and 'country' with 'Unknown'
df1['rating'] = df['rating'].fillna('Unknown')
df1['country'] = df['country'].fillna('Unknown')
```

```python
# Checking Data after cleaning
df1.shape
```

```python
# Checking null values after cleaning , it is None as values are either removed or filled up
df1.isnull().sum()
```

## Genre Analysis

```python
# From here we use a particular column 'Listed_in' and create a 'genre_list' new column to seperate genres for each title
df1['genre_list'] = df1['listed_in'].str.split(',')
```

```python
# Explode the dataframe to have one genre per row
df1 = df1.explode('genre_list')
```

```python
# Strip spaces from genre names
df1['genre_list'] = df1['genre_list'].str.strip(' ')
df1
```

```python
# Count the most common genres
genres_count = df1['genre_list'].value_counts()
genres_count.head(10)
```

## Visualization

```python
# Resetting the index to a serial from genre_list
top_genres = genres_count.head(10).reset_index()
top_genres
```

```python
# Bar plot of the top 10 genres using Seaborn
plt.figure(figsize=(10,6))
sns.barplot(x = 'genre_list', y = 'count', data = top_genres, palette = 'viridis')
plt.title("Top 10 Genres of Amazon Prime")
plt.xlabel('genre_list')
plt.ylabel('Number of Titles')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
